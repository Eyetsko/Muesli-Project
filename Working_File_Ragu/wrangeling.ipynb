```python
#Create df where value "on truck scan date" & "arrival_scan_day" exist
df_log = df_raw.dropna(subset=["On Truck Scan Date", "Arrival Scan Date"])
```

```python
df_log.shape
```

```python
df_log.dtypes
```

```python
cols = ["On Truck Scan Date", "Arrival Scan Date",'Order Date_x','Order Date_y', 'Ready to Ship Date','Pickup Date' ]
df_log[cols] = df_log[cols].apply(pd.to_datetime)
df_log
```

```python
df_log.loc[:, "delivery_time_days"] = (df_log["Arrival Scan Date"] - df_log["On Truck Scan Date"]).dt.days
df_log.shape
```

```python
df_log.loc[:, "total_delivery_time_days"] = (df_log["Arrival Scan Date"] - df_log["Order Date_x"]).dt.days
df_log.shape
```

```python
df_log['Order Date_x'] = pd.to_datetime(df_log["Order Date_x"])
df_log["order_Month"] = df_log["Order Date_x"].dt.strftime("%B")

df_log
```

```python
df_log['Order_YearMonth'] = df_log['Order Date_x'].dt.to_period('M')
```

```python
df_log['order_weekday'] = df_log['Order Date_x'].dt.day_name()
df_log['on_truck_weekday'] = df_log['On Truck Scan Date'].dt.day_name()
```

```python
df_log.loc[:, "warehouse_time_days"] = (df_log["total_delivery_time_days"] - df_log["delivery_time_days"])
```
